```python
import pandas as pd

df = pd.read_csv('datos_ML.csv')
```

Estructura del gráfico: El gráfico es una representación tridimensional de puntos en un espacio 3D, donde cada punto representa una propiedad. Los ejes del gráfico son:

Eje X: Representa la superficie construida de cada propiedad.

Eje Y: Representa el número de habitaciones en cada propiedad.

Eje Z: Representa el precio de cada propiedad.

Colorización: Las propiedades se colorean según su ubicación en el distrito. Esto permite visualizar cómo las propiedades se distribuyen en diferentes distritos.

Filtrado de datos: El gráfico solo muestra las propiedades que tienen un precio inferior a 1.000.000. Esto se logra mediante la condición df['Precio'] < 1000000 en la línea df.loc[df['Precio'] < 1000000].

Configuración del gráfico: El gráfico tiene un ancho y alto de 1000 pixeles, lo que permite una visualización clara y detallada de los datos.

```python
import plotly.express as px 
import nbformat

fig = px.scatter_3d(df.loc[df['Precio'] < 1000000], x = 'Superficie construida', y = 'Habitaciones', z = 'Precio', color='Distrito')
fig.update_layout(width=1000, height=1000)
fig.show()
```

Este gráfico es una representación de distribución de precios de propiedades en diferentes distritos. Aquí hay lo que se ve:

Eje X: Representa los distritos.

Eje Y: Representa los precios de las propiedades.

Boxplots: Cada boxplot muestra la distribución de los precios en un distrito. La línea central es la media, la parte superior es el 75% de los datos y la parte inferior es el 25%.

Puntos: Todos los datos individuales se muestran como puntos en el gráfico, lo que permite ver la distribución de los precios en cada distrito.

Filtrado: Solo se muestran las propiedades que cuestan menos de 1.000.000 de dólares.

Configuración del gráfico: El gráfico tiene un ancho y alto de 1000 pixeles, lo que permite una visualización clara y detallada de los datos.


```python
fig = px.box(df.loc[df['Precio'] < 1000000], x = 'Distrito', y = 'Precio', points = 'all')
fig.update_layout(width=1000, height=1000)
fig.show()
```
